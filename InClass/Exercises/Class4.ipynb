```python
#8 Create a new column in penguins called bill_ratio for the penguin dataset that contains the 
#  ratio of bill length to bill depth (in otherwords bill length divided by bill depth). Print out this column.

bill_lengths = penguins["bill_length_mm"]
bill_depths = penguins["bill_depth_mm"]

bill_ratios = []

for i in range(len(bill_lengths)):
    bill_ratios.append(bill_lengths[i]/bill_depths[i])
    
penguins['bill_ratios'] = bill_ratios

print(penguins['bill_ratios'])
```

```python
#9 Create a new data frame called lil_boys that only contains penguins that have a body mass less than 4000 grams. 
#  Print the first 4 rows of this dataframe.

lil_boy = penguins['body_mass_g'] < 4000
lil_boys = penguins.loc[lil_boy]
print(len(lil_boys))
lil_boys.head(4)
```

```python
#10 What is the mean bill length for each species of penguin? 
#   Store the means in the variables bl_Adelie, bl_Chinstrap and bl_Gentoo respectively. Print them out.

isAdelie = penguins['species'] == "Adelie"
isChinstrap = penguins['species'] == "Chinstrap"
isGentoo = penguins['species'] == "Gentoo"

adelies = penguins.loc[isAdelie]
chinstraps = penguins.loc[isChinstrap]
gentoos = penguins.loc[isGentoo]

bl_Adelie = adelies['bill_length_mm'].mean()
bl_Chinstrap = chinstraps['bill_length_mm'].mean()
bl_Gentoo = gentoos['bill_length_mm'].mean()

print("Adelie mean bill length: " + str(bl_Adelie))
print("Chinstrap mean bill length: " + str(bl_Chinstrap))
print("Gentoo mean bill length: " + str(bl_Gentoo))
```

```python
#11 Which class from the titanic dataset had the highest proportion of surviors? 
#   Print out the proportion of people who survived for each class.

is_first_class = titanic['class'] == "First"
is_second_class = titanic['class'] == "Second"
is_third_class = titanic['class'] == "Third"

first_class = titanic.loc[is_first_class]
second_class = titanic.loc[is_second_class]
third_class = titanic.loc[is_third_class]

first_survived = 0
second_survived = 0
third_survived = 0

for p in first_class['survived']:
    first_survived = first_survived + p
    
for p in second_class['survived']:
    second_survived = second_survived + p
    
for p in third_class['survived']:
    third_survived = third_survived + p
    
# First class had the highest proportion of survivors
    
print(first_survived / len(first_class.index))
print(second_survived / len(second_class.index))
print(third_survived / len(third_class.index))
```
